```python
X = ['a'] * 100
Y = ['b'] * 100
```

```python
#用于第一次创建文件并存储
import pandas as pd
file_path = 'f:/csv_file/lianxi.csv'
dfa = pd.DataFrame(data={'X':X, 'Y':Y})
dfa.to_csv(file_path,index=False)
```

```python
#读取数据
dfa = pd.read_csv(file_path)
dfa
```

```python
#添加数据
C = ['c'] * 100
D = ['d'] * 100
dfa.loc[:len(C)-1,'C'] = C
dfa.loc[:len(D)-1,'D'] = D
dfa.to_csv(file_path,index=False)
```

```python
dfa
```

```python
#删除数据
dfa.drop(['C'],axis=1,inplace=True)
```

```python
dfa
```

```python
#添加不同行的数据
C = ['c'] * 90
dfa.loc[:len(C)-1,'C'] = C
dfa.to_csv(file_path,index=False)
dfa
```

```python
#寻找某一列最小值及其行所在位置
E = range(100)
dfa.loc[:len(E)-1,'E'] = E
dfa.to_csv(file_path,index=False)
print(dfa[dfa['E'] == dfa.loc[:,'E'].min()].index.values[0])
dfa[dfa['E'] == dfa.loc[:,'E'].min()]
```

```python
#寻找满足任何条件的值
import numpy as np
F = np.random.randint(100,size=100)
dfa.loc[:len(F)-1,'F'] = F
dfa.to_csv(file_path,index=False)
dfa['F'][dfa['F'].diff() <= 10]
```

```python
#修改数据
dfa.loc[dfa['F'] < 50,'F'] = np.nan
dfa.to_csv(file_path,index=None)
dfa
```
